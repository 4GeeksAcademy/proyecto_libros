```python
import pandas as pd
import requests
import time
def get_google_books_info(title):
    """Devuelve rating, numRatings, coverImg desde Google Books API."""
    try:
        url = f"https://www.googleapis.com/books/v1/volumes?q=intitle:{title}"
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            if "items" in data and len(data["items"]) > 0:
                volume = data["items"][0]["volumeInfo"]
                return {
                    "rating": volume.get("averageRating"),
                    "numRatings": volume.get("ratingsCount"),
                    "coverImg": volume.get("imageLinks", {}).get("thumbnail"),
                }
    except:
        pass
    return {}

def get_openlibrary_info(title):
    """Devuelve setting, characters, awards desde OpenLibrary API."""
    try:
        url = f"https://openlibrary.org/search.json?title={title}"
        res = requests.get(url)
        if res.status_code == 200:
            data = res.json()
            if data['docs']:
                doc = data['docs'][0]
                return {
                    "setting": ", ".join(doc.get("place", [])) if "place" in doc else None,
                    "characters": ", ".join(doc.get("person", [])) if "person" in doc else None,
                    "awards": ", ".join(doc.get("awards", [])) if "awards" in doc else None,
                }
    except:
        pass
    return {}

# --- RELLENO DE CAMPOS FALTANTES ---

def completar_info_externa(df, max_libros=100):
    procesados = 0

    for idx, row in df.iterrows():
        if procesados >= max_libros:
            break

        title = row['title']
        update_needed = (
            pd.isna(row['rating']) or
            pd.isna(row['numRatings']) or
            pd.isna(row['coverImg']) or
            pd.isna(row['setting']) or
            pd.isna(row['characters']) or
            pd.isna(row['awards'])
        )

        if not update_needed:
            continue

        print(f"Buscando: {title}")
        info_google = get_google_books_info(title)
        info_openlib = get_openlibrary_info(title)

        if pd.isna(row['rating']) and info_google.get('rating') is not None:
            df.at[idx, 'rating'] = info_google['rating']
        if pd.isna(row['numRatings']) and info_google.get('numRatings') is not None:
            df.at[idx, 'numRatings'] = info_google['numRatings']
        if pd.isna(row['coverImg']) and info_google.get('coverImg') is not None:
            df.at[idx, 'coverImg'] = info_google['coverImg']

        if pd.isna(row['setting']) and info_openlib.get('setting') is not None:
            df.at[idx, 'setting'] = info_openlib['setting']
        if pd.isna(row['characters']) and info_openlib.get('characters') is not None:
            df.at[idx, 'characters'] = info_openlib['characters']
        if pd.isna(row['awards']) and info_openlib.get('awards') is not None:
            df.at[idx, 'awards'] = info_openlib['awards']

        procesados += 1
        time.sleep(1.1)  # Evita ser bloqueado por las APIs

    print(f"Completados {procesados} libros.")
    return df
```

```python
df = completar_info_externa(df, max_libros=74000)
```

```python
df.to_csv("../data/full_data/libros_completados.csv", index=False)
```
